## 480. Sliding Window Median
`Hard`

The median is the middle value in an ordered integer list. If the size of the list is even, there is no middle value. So the median is the mean of the two middle values.

* For examples, if `arr = [2,3,4]`, the median is `3`.
* For examples, if `arr = [1,2,3,4]`, the median is `(2 + 3) / 2 = 2.5`.
You are given an integer array `nums` and an integer `k`. There is a sliding window of size k which is moving from the very left of the array to the very right. You can only see the `k` numbers in the window. Each time the sliding window moves right by one position.

Return the median array for each window in the original array. Answers within `10-5` of the actual value will be accepted.
 

**Example 1:**

**Input**: nums = [1,3,-1,-3,5,3,6,7], k = 3

**Output**: [1.00000,-1.00000,-1.00000,3.00000,5.00000,6.00000]

**Explanation**: 

|      Window position       |      Median      |
|----------------------------|------------------|
| `[1  3  -1]` -3  5  3  6  7  |  1               |
| 1 `[3  -1  -3]` 5  3  6  7   |    -1            |
| 1  3 `[-1  -3  5]` 3  6  7   | -1               |
| 1  3  -1 `[-3  5  3]` 6  7   | 3                |    
| 1  3  -1  -3 `[5  3  6]` 7   | 5                |
| 1  3  -1  -3  5 `[3  6  7]`  | 6                |


**Example 2:**

**Input**: nums = [1,2,3,4,2,3,1,4,2], k = 3

**Output**: [2.00000,3.00000,3.00000,3.00000,2.00000,3.00000,2.00000]
 

```python
class Solution:
    def medianSlidingWindow(self, nums: List[int], k: int) -> List[float]:
        res = []

        # max heap
        small = []
        heapify(small)
        # min heap
        large = []
        heapify(large)
        # to keep track the out of window numbers
        out_nums = {}

        i = 0
        # spread the first k numbers in the max heap
        for i in range(k):
            heappush(small, -nums[i])

        i += 1
        # push half of the max heap in min heap
        for j in range(k//2):
            heappush(large, -heappop(small))
        while True:
            # append the median for the window
            if (k & 1) == 1:
                res.append(-small[0])
            else:
                res.append((large[0] - small[0]) / 2)

            if i >= len(nums): break
            balance = 0
            out = nums[i - k]
            inn = nums[i]
            i += 1
            # update the balance
            if out <= -small[0]:
                balance -= 1
            else:
                balance += 1

            # update the out_nums hash
            if out in out_nums:
                out_nums[out] += 1
            else:
                out_nums[out] = 1

            # insert the new number to the window 'heaps'
            if small and inn <= -small[0]:
                heappush(small, -inn)
                balance += 1
            else:
                heappush(large, inn)
                balance -= 1

            # balance the tow heaps 
            if balance < 0:
                heappush(small, -heappop(large))
                balance += 1
            if balance > 0:
                heappush(large, -heappop(small))
                balance -= 1
                
            while small and -small[0] in out_nums and out_nums[-small[0]] > 0:
                out_nums[-small[0]] -= 1
                heappop(small)

            while large and large[0] in out_nums and out_nums[large[0]] > 0:
                out_nums[large[0]] -= 1
                heappop(large)
        
        return res
```
