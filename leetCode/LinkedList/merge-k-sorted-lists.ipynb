
# 23. Merge k Sorted Lists

You are given an array of `k` linked-lists `lists`, each linked-list is sorted in ascending order.

Merge all the linked-lists into one sorted linked-list and return it.

 
-------------
**Example 1:**

**Input:** `lists = [[1,4,5],[1,3,4],[2,6]]`

**Output:** `[1,1,2,3,4,4,5,6]`

**Explanation:** `The linked-lists are:`

`[`

`  1->4->5,`

`  1->3->4,`

`  2->6`

`]`

`merging them into one sorted list:`
`1->1->2->3->4->4->5->6`

-------------

**Example 2:**

**Input:** `lists = []`

**Output:** `[]`

-------------
**Example 3:**

**Input:** `lists = [[]]`

**Output:** `[]`
 
-------------

**Constraints:**

`k == lists.length`

`0 <= k <= 104`

`0 <= lists[i].length <= 500`

`-104 <= lists[i][j] <= 104`

`lists[i] is sorted in ascending order.`

`The sum of lists[i].length will not exceed 104.`

```python
# sol 1: brut O(n * k)
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def mergeKLists(self, lists: List[Optional[ListNode]]) -> Optional[ListNode]:
        if len(lists) == 0:
            return None
        listElements = []
        for i in range(len(lists)):
            nodeList = lists[i]
            while nodeList:
                listElements.append(nodeList.val)
                nodeList = nodeList.next
        if len(listElements) == 0:
            return None
        
        listElements.sort()    
        resultList = ListNode(listElements[0])
        current = resultList
        
        for i in range(1,len(listElements)):
            current.next = ListNode(listElements[i])
            current = current.next
            
        return resultList 
        
```

```python
# sol 2: O(n * log(k))
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def mergeKLists(self, lists: List[Optional[ListNode]]) -> Optional[ListNode]:
        if len(lists) == 0:return None
        for i in range(len(lists) - 1):
            lists[i + 1] = self.mergeTowLists(lists[i], lists[i + 1])

        return lists[-1]

    def mergeTowLists(self, l1, l2):
        dummy = ListNode()
        cur = dummy
        while l1 and l2:
            if l1.val <= l2.val:
                cur.next = l1
                l1 = l1.next
            else:
                cur.next = l2
                l2 = l2.next
            cur = cur.next
        cur.next = l1 or l2

        return dummy.next
```
